'''sql 练习用表格创建

1、对公活期表，共2000行，共包含字段

    *cust_id_sae 客户号

    *Mbal1rmb_sae  月均rmb余额

    *org_cd_sae 机构号


2、对公定期表，共1000行，包含字段
  
      *cust_id_tde 客户号

      *Dbalrmb_tde  日均rmb余额

      *Mbalrmb_tde  月均rmb余额

      *Mbalusd_tde  月均美元余额

      *org_cd_tde 机构号
      

'''

```python
import pandas as pd
import random
```

```python
# 随机创建整数函数（创建客户号、机构号）
def random_no(start,stop,length):
    if length>=0:
        length=int(length)
        start, stop = (int(start), int(stop)) if start <= stop else (int(stop), int(start))
    
    randomNo_list = []
    
    for i in range(length):
        randomNo_list.append(random.randint(start, stop))
    return randomNo_list
```

```python
# 随机创建浮点数函数（创建各种余额）
def random_float(start,stop,length):
    if length>=0:
        length=int(length)
        start, stop = (int(start), int(stop)) if start <= stop else (int(stop), int(start))
    
    randomNo_list = []
    
    for i in range(length):
        randomNo_list.append(round(random.uniform(start, stop), 2))
    return randomNo_list
```

```python
#定义客户信息表中的客户号列表
custID_list = random_no(1000,2000,200)
print(custID_list)
```

```python
# 创建机构号列表
orgID_list = random_no(100,200,30)
print(orgID_list)
```

```python
# 创建机构名称
#coding=utf-8
orgName_list = ['南京分行','上海分行','北京分行','杭州分行','无锡分行','江北新区分行','宁波分行','常州分行','苏州分行',
                '广州分行','城南分行','城东分行','城西分行','连云港分行','南通分行','徐州分行','合肥分行','曹县分行',
                '泰州分行','淮安分行','扬州分行','宿迁分行','镇江分行','盐城分行','华西村分行','石家庄分行','洛阳分行',
               '济南分行','青岛分行','昆山分行']
```

```python
org_tuple = list(zip(orgID_list, orgName_list))
print(org_tuple)
```

### 创建对公活期表

```python
# 创建活期对应客户
cust_choose = custID_list[1:150]
sae_custNo = []
for i in range(2000):
    sae_custNo.append(random.sample(cust_choose,1)[0])
    
print(sae_custNo[-1])
```

```python
# 创建活期月均余额
mbal_sae_ls = random_float(10000,200000,2000)
print(mbal_sae_ls)
```

```python
# 创建活期余额对应机构信息
org_sae = []
length = 2000
for i in range(length):
    org_sae.append(random.sample(org_tuple,1)[0])

print(org_sae[-10:-1])
```

```python
# 拆分机构号与机构名
org_cd_sae, org_name_sae = zip(*org_sae)   # 快捷技巧
    
print(org_cd_sae[-1], org_name_sae[-1])
```

```python
# 转化为数据框
sae_dict= {"custID":sae_custNo,
   "m_avg_bal":mbal_sae_ls,
  "org_cd":org_cd_sae,
  "org_name":org_name_sae}

sae_frame =pd.DataFrame(sae_dict)

sae_frame.head()
```

```python
# 活期表另存为csv文件
sae_frame.to_csv('sae_table.csv', encoding='utf_8_sig', index=0)
```

### 创建对公定期表

```python
# 创建定期对应客户
cust_choose2 = custID_list[120:199]
tde_custNo = []
for i in range(1000):
    tde_custNo.append(random.sample(cust_choose2,1)[0])
    
print(tde_custNo[-1])
```

```python
# 创建定期月均人民币余额
mbal_tde_rmb = random_float(10000,800000,1000)
print(mbal_tde_rmb[0:10])
```

```python
# 创建定期月均美元余额
mbal_tde_usd = random_float(6000,50000,1000)
print(mbal_tde_usd[0:10])
```

```python
# 创建定期机构信息
org_tde = []
for i in range(1000):
    org_tde.append(random.sample(org_tuple,1)[0])

print(org_tde[-1])
```

```python
org_cd_tde, org_name_tde = zip(*org_tde)   # 拆分机构号机构名
    
print(org_cd_tde[-1], org_name_tde[-1])
```

```python
# 转化为数据框
tde_dict= {"custID":tde_custNo,
   "m_avg_rmb":mbal_tde_rmb,
  "m_avg_usd":mbal_tde_usd,
  "org_cd":org_cd_tde,
  "org_name":org_name_tde}

tde_frame =pd.DataFrame(tde_dict)

tde_frame.head()
```

```python
# 定期表另存为csv文件
tde_frame.to_csv('tde_table.csv', encoding='utf_8_sig', index=0)
```
